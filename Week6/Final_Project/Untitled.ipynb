```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
veri = pd.read_csv('companies.csv')
```

```python
satir_sayisi, sutun_sayisi = veri.shape
print("Boyut: Satır {}, Sütun {}".format(satir_sayisi, sutun_sayisi))

sutun_isimleri = veri.columns
print("Sütun İsimleri: {}".format(sutun_isimleri))
```

```python
country_counts = veri["country"].value_counts()
en_cok_10_ulke = country_counts.head(20)

plt.figure(figsize=(12, 6))
ax = en_cok_10_ulke.plot(kind='bar', color='skyblue')
plt.title("Ülke Dağılımı - En Çok Adı Geçen 20 Ülke")
plt.xlabel("Ülke")
plt.ylabel("Sayı")
plt.xticks(rotation=45)
plt.yticks(range(0, en_cok_10_ulke.max() + 1, 50))

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')

plt.show()
```

```python
industry_counts = veri['industry'].value_counts()

labels = industry_counts.index
sizes = industry_counts.values

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Endüstri Türü Dağılımı')
plt.show()
```

```python
country_counts = veri['country'].value_counts(normalize=True) * 100

filtered_countries = country_counts[country_counts > 1]

labels = filtered_countries.index
sizes = filtered_countries.values

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Ülke Dağılımı (Yüzdelik Oranı %1 ve Üzeri)')
plt.show()
```

```python
exited_countries = veri[veri["action"] == "suspend operations in Russia"]

exited_country_counts = exited_countries["country"].value_counts()

top_10_exited_countries = exited_country_counts.head(10)

plt.pie(top_10_exited_countries, labels=top_10_exited_countries.index, autopct="%1.1f%%")
plt.title("Rusya'dan Ayrılan En Fazla 10 Ülke")
plt.show()
```

```python
turkey_data = veri[veri['country'] == 'Turkey']

industry_counts = turkey_data['industry'].value_counts()

plt.figure(figsize=(12, 6))
industry_counts.plot(kind='bar')
plt.xlabel('Endüstri')
plt.ylabel('Şirket Sayısı')
plt.title('Türkiye\'deki Şirketlerin Endüstri Dağılımı')
plt.xticks(rotation=45)
plt.show()
```

```python
exited_countries = veri[veri["action"] == "Fully exit Russia after selling its operations"]
operating_countries = veri[veri["action"] == "still operating in Russia"]

exited_country_count = exited_countries.shape[0]
operating_country_count = operating_countries.shape[0]

total_country_count = veri.shape[0]

exited_ratio = exited_country_count / total_country_count
operating_ratio = operating_country_count / total_country_count

plt.pie([exited_ratio, operating_ratio], labels=["Rusya'dan Ayrılanlar", "Rusya'da Kalanlar"], autopct='%1.1f%%')
plt.title("Rusya'dan Ayrılan ve Rusya'da Kalan Şirketlerin Oranı")
plt.show()
```

```python
business_as_usual_data = veri[veri['action'] == 'Business as usual']

country_counts = business_as_usual_data['country'].value_counts()

plt.figure(figsize=(12, 6))
country_counts.plot(kind='bar')
plt.xlabel('Ülke')
plt.ylabel('Şirket Sayısı')
plt.title('"Business as Usual" Eylemiyle Devam Eden Şirket Sayısı (Ülkeler)')
plt.xticks(rotation=45)
plt.show()
```

```python
industrials_data = veri[veri['industry'] == 'Financials']

country_counts = industrials_data['country'].value_counts()

plt.figure(figsize=(12, 6))
country_counts.plot(kind='bar')
plt.xlabel('Ülke')
plt.ylabel('Şirket Sayısı')
plt.title('"Financials" Endüstrisindeki Şirket Sayısı (Ülkeler)')
plt.xticks(rotation=45)
plt.show()
```

```python
france_companies = veri[veri['country'] == 'France']

industry_counts = france_companies['industry'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Daireyi tam daire olarak ayarlayın
plt.title("Fransa'da Endüstrilere Göre Şirket Dağılımı")
plt.show()
```

```python
still_operating_companies = veri[veri['action'].str.contains('still operating in Russia|still flying to Russia')]

total_companies = veri.shape[0]

# "still operating in Russia" veya "still flying to Russia" olan şirketlerin sayısını hesaplayın
still_operating_count = still_operating_companies.shape[0]

# Oranı hesaplayın
operating_ratio = still_operating_count / total_companies
flying_ratio = 1 - operating_ratio

# Pasta grafiği oluşturun
labels = ['Still Operating in Russia', 'Still Flying to Russia']
sizes = [operating_ratio, flying_ratio]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Daireyi tam daire olarak ayarlayın
plt.title("Rusya'da Faaliyette Olan ve Rusya'ya Uçan Şirketlerin Oranı")
plt.show()
```

```python
china_companies_count = veri[veri['country'] == 'China'].shape[0]
germany_companies_count = veri[veri['country'] == 'Germany'].shape[0]
france_companies_count = veri[veri['country'] == 'France'].shape[0]

# Oranları hesaplayın
total_companies_count = china_companies_count + germany_companies_count + france_companies_count
china_ratio = china_companies_count / total_companies_count
germany_ratio = germany_companies_count / total_companies_count
france_ratio = france_companies_count / total_companies_count

# Pasta grafiği oluşturun
labels = ['China', 'Germany', 'France']
sizes = [china_ratio, germany_ratio, france_ratio]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Daireyi tam daire olarak ayarlayın
plt.title("Çin, Almanya ve Fransa'daki Şirketlerin Oranı")
plt.show()
```

```python
suspend_shipments_count = veri[veri['action'] == 'Suspend all shipments to Russia'].shape[0]
still_operating_count = veri[veri['action'] == 'still operating in Russia'].shape[0]
suspend_operations_count = veri[veri['action'] == 'suspend operations in Russia'].shape[0]
business_as_usual_count = veri[veri['action'] == 'Business as usual'].shape[0]

labels = ['Suspend all shipments to Russia', 'Still Operating in Russia', 'Suspend operations in Russia', 'Business as usual']
sizes = [suspend_shipments_count, still_operating_count, suspend_operations_count, business_as_usual_count]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title("Rusya'da Gönderileri Durduran, Faaliyette Olan, Faaliyetleri Durduran ve Normal Şekilde Devam Eden Şirketlerin Oranları")
plt.show()
```
